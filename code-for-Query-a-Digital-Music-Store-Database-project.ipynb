# Query 1 - The top 5 countries have the most Invoices:

```python
SELECT BillingCountry AS Billing_Country,COUNT(*) AS Invoices 
FROM Invoice 
GROUP BY Billing_Country 
ORDER BY Invoices DESC
LIMIT 5 ;
```

# Query 2 - The top 5 band Who is writing the Jazz music:


```python
SELECT ar.name ,count(*) as total_track, a.artistid
from genre g 
join  Track t
ON t.genreid=g.genreid
join Album a
ON t.albumid = a.albumid
JOIN Artist ar
on a.artistid = ar.artistid
WHERE g.name = 'Jazz'
GROUP by a.artistid
ORDER by total_track DESC
LIMIT 5 ;
```

# Query 3 -  The top 3 artist has earned the most according to the Invoices:

```python
SELECT sum(i.UnitPrice * i.Quantity) as 'Artist_Total', ar.Name as 'Artis_tName',c.customerid,c.firstname
FROM Artist ar
join Album a
on ar.artistid=a.artistid
JOIN Track t
on a.albumid=t.albumid
join InvoiceLine i
on  t.trackid=i.trackid
JOIN Invoice inv
on inv.invoiceid=i.invoiceid
join Customer c
on inv.customerid=c.customerid
GROUP BY Artis_tName
ORDER BY Artist_Total desc
LIMIT 3;
```

# Query 4 - The top 3  countries has the best customers:

```python
SELECT billingcountry AS billingcountry,SUM(Total)  AS InvoiceDollars 
FROM Invoice 
GROUP BY billingcountry
ORDER BY InvoiceDollars DESC
LIMIT 3;
```
